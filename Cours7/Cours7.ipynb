### La méthode `plot` avec les objets Pandas

Si on désire faire des graphiques avec des données contenues dans un objet Pandas (Series ou DataFrame). On peut utiliser directement la méthode `plot()`.

Voici le lien vers la documentation pour tous les types de graphiques supportés: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html

```python
# Création d'un DataFrame fictif
data = {
    'Année': [2018, 2018, 2019, 2019, 2020, 2020, 2021, 2021],
    'Produit': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Ventes': [100, 150, 120, 180, 200, 250, 220, 280]
}

df = pd.DataFrame(data)

# Création de la table pivot
pivot_table = df.pivot_table(index='Année', columns='Produit', values='Ventes', aggfunc='sum')

# Création du graphique à partir du tableau croisé dynamique
# axe x: Index
# axe y: Values
# légende: columns
pivot_table.plot()
plt.xlabel('Année')
plt.ylabel('Ventes')
plt.title('Ventes par Produit et Année')
plt.legend(title='Produit')
plt.show()
```

### Bubble Charts

Avec la fonction `plt.scatter`, les propriétés de chaque point individuel (taille, couleur de remplissage, couleur des bords, etc.) peuvent être contrôlées individuellement ou en fonction des données contenus dans une variable.

Prenons, par exemple, le jeu de données Iris de la librairie Scikit-Learn où chacune des observations de ce jeu de donnée est une des 3 espèces de fleurs suivantes:

* setosa
* versicolor
* virginica

Les données concernent la taille mesurée de ces pétales (petals) et de ces feuilles (sepals) pour chacune des observations.

Avec l'exemple ci-dessous, on présente quatre variables en un seul graphique:
* La position en x (features[0]) représente la longueur de la feuille. 
* La position en y (features[1]) représente la largeur de la feuille. 
* La taille des points (s=100*features[3]) représente la largeur des pétales.
* La couleur des points (c=iris.target) nous indique l'espèce de fleur.

L'argument `cmap` permet de relier les données numériques dans l'argument `c` à des couleurs.
Voici le lien vers les différentes `cmap`acceptées avec les fonctions de Matplotlib: https://matplotlib.org/stable/users/explain/colors/colormaps.html

L'argument `alpha` nous permet d'ajuster la transparence des points.

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

fig = plt.figure()
ax = plt.axes()

ax.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_title('Sepal length in function of sepal width')
```

### Histogrammes

Un histogramme est un type de représentation graphique des données qui montre la répartition de fréquences ou de densités de valeurs dans un ensemble donné. L'axe horizontal de l'histogramme représente les valeurs possibles du jeu de données, regroupées en intervalles appelés "classes" ou "bin". L'axe vertical représente la fréquence ou la densité des occurrences de ces valeurs dans chaque classe.

La fonction `plt.hist` permet de créer un (des) histogrammes. L'argument `bins` nous permet de définir en combien de classe on souhaite diviser nos valeurs.

```python
rng = np.random.RandomState(0)
data = rng.randn(1000)
plt.hist(data, bins=30,  alpha=0.5,
         color='steelblue')
```

```python
rng = np.random.RandomState(0)
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)
fig = plt.figure()
ax = plt.axes()



ax.hist(x1,alpha=0.8, bins=40)
ax.hist(x2,alpha=0.8, bins=40)
ax.hist(x3,alpha=0.8, bins=40)
```

### Diagramme de dispersion hexagonal (*hexagon bin plot*)
Un graphique Hexagon Bin Plot (ou diagramme de dispersion hexagonal) est un type de représentation graphique utilisé pour visualiser la distribution conjointe de deux variables continues. Ce graphique est particulièrement utile lorsque les données sont nombreuses et qu'il est difficile de voir les détails dans un nuage de points traditionnel. Il combine les avantages d'un nuage de points classique et d'un histogramme en regroupant les données en hexagones, ce qui permet d'obtenir une vue d'ensemble des densités de points dans différentes régions de l'espace des variables.

L'espace des variables est divisé en une grille hexagonale. Chaque hexagone représente une petite région de l'espace délimitée par ses côtés. Pour chaque hexagone, le nombre d'observations qui tombent dans cet hexagone est compté. Plus il y a d'observations dans un hexagone, plus il sera sombre sur le graphique.

```python
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
#plus le gridsize est petit, il y aura moins d'hexagone donc plus de points regroupé ensemble
plt.hexbin(iris['sepal_length'], iris['sepal_width'], gridsize=30, cmap='Blues')
cb = plt.colorbar(label='Nb obsv.')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
```

Le diagramme en bâtons empilés est utilisée pour représentée plusieurs variables catégoriques et numériques ensemble. Ci-dessous, nous avons les ventes par produit et par magasin.
Il existe plusieurs façon de créer un diagramme en bâtons empilés. Ici, on a utilisé la méthode `.plot.bar()` de Pandas en combinaison avec les fonctions de Matplotlib pour annoter le graphique (noms des axes, légende etc.)

```python
import pandas as pd
Vente_A = [100, 200]
Vente_B = [50, 250]
index = ['Magasin_1','Magasin_2']
df = pd.DataFrame({'Produit A': Vente_A,
                   'Produit B': Vente_B}, index=index)
ax = df.plot.bar(stacked=True)
plt.legend(loc="upper left", ncol=2)
plt.xlabel("Magasin")
plt.ylabel("Ventes")
plt.title("Ventes par magasin et produit")
for n, x in enumerate([*df.index.values]):
    for (vente, y_loc) in zip(df.loc[x], df.loc[x].cumsum()):
        total=df.loc[x].sum()
                       
        plt.text(x=n-0.1,
                 y=y_loc,
                 s=f'{np.round(vente/total*100,2)}%', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")
```

Le premier boucle énumère et parcourt les index du *DataFrame*. Nous utiliserons les valeurs de l'énumération (0, 1) comme position sur l'axe des étiquettes de données. En Python, l'opérateur * est utilisé pour décompresser des listes, des tuples ou d'autres structures de données itérables. Lorsqu'on utilise * devant une liste, il extrait les éléments de la liste et les passe en tant qu'arguments individuels à une fonction ou les convertit en des éléments distincts dans une nouvelle liste ou un nouvel objet itérable.

```python
for n, x in enumerate([*df.index.values]):
    print(n,x)
```

La fonction `text()` prend en argument les éléments suivants dans cet exemple:
* la position du texte sur l'axe des x (x)
* la position du texte sur l'axe des y (y)
* le texte (s)
* la couleur du texte (color)
* la grosseur du texte (fontsize)

Il existe d'autres arguments et je vous invite à consulter la documentation:https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html

On doit modifier l'emplacement des valeurs pour la 2e valeur car on veut que ce soit situé en haut de la deuxième section du graphique. On modifie la deuxième boucle afin de calculer la somme cumulative comme position sur l'axe des y.

La fonction `zip()` en Python est utilisée pour combiner des éléments de plusieurs séquences (comme des listes, des tuples ou d'autres objets itérables) et les agréger en une seule séquence. Elle crée un nouvel objet itérable en associant les éléments correspondants des séquences fournies.

```python
for n, x in enumerate([*df.index.values]):
    for (vente, y_loc) in zip(df.loc[x], df.loc[x].cumsum()):
        print(n,x,vente,y_loc)
```

On peut également créer un diagramme en bâtons groupés plutôt qu'empilés. Dans ce cas on peut utiliser, la méthode `containers()` pour inscrire les valeurs dans les bâtons.

### Heatmap (matrice de corrélation)
La matrice de corrélation est la matrice des coefficients de corrélation calculés sur plusieurs variables prises deux à deux. Elle permet de regrouper dans un seul tableau toutes les relations entre les variables pour évaluer la dépendance entre plusieurs variables en même temps. Cette matrice est symétrique.
Pour obtenir la matrice de corrélation, on peut utiliser la fonction `corr()` de la librairie Pandas. Ensuite, avec cette matrice de corrélation, on peut utiliser la fonction `heatmap`de la librairie Seaborn

```python
data=iris.loc[:,'sepal_length':'petal_width']
corr=data.corr()
sns.heatmap(corr, annot=True)
```

Les valeurs près de 1 signifie une corrélation positive et les valeurs près de -1 signifie une corrélation négative. Ici, on voit qu'il existe une corrélation (0.96) entre la longueur de la pétale et sa largeur.

## Introduction aux statistiques déductives et inférences

### Diverses formes de distribution 

#### Distribution normale

La distribution normale, également connue sous le nom de distribution gaussienne, est l'une des distributions de probabilité les plus importantes en statistiques. Elle est caractérisée par sa forme en cloche symétrique. Dans une distribution normale, la moyenne (la moyenne arithmétique), la médiane (la valeur du milieu lorsque les données sont triées) et le mode (la valeur la plus fréquente) sont tous égaux et situés au point le plus élevé de la cloche.

La distribution normale est complètement déterminée par deux paramètres : la moyenne (µ) et l'écart-type (σ). La moyenne indique le centre de la distribution, tandis que l'écart-type mesure la dispersion des valeurs par rapport à la moyenne. Plus l'écart-type est grand, plus la distribution est étalée.

```python
from scipy.stats import norm
# On fixe la moyenne et l'écart-type
mu, sigma = 0, 1 

# On génère des chiffres aléatoires tirés d'une distribution normale avec les paramètres définis plus haut
rng = np.random.RandomState(0)
data = rng.normal(mu, sigma, 1000)

# Histogramme
plt.hist(data,
         bins=50,
         density=True, #y_value= counts / (sum(counts) * bins_width)
         alpha=0.6,
         color='g')
#Probability Density function (PDF): calcule la probabilité d'observée une certaine valeur.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'black', linewidth=2)

plt.title("Distribution Normale")

plt.show()
```

#### Distribution binomiale

La distribution binomiale est un modèle de probabilité discret utilisé pour décrire le nombre de succès dans un nombre fixe d'essais ou d'expériences indépendantes, chacun ayant deux résultats possibles : succès ou échec. Chaque essai est indépendant des autres, et la probabilité de succès reste constante pour chaque essai. La distribution binomiale est largement utilisée dans les statistiques pour modéliser des événements binaires, comme le lancer d'une pièce de monnaie (face ou pile).
La distribution binomiale est définie par deux paramètres : le nombre d'essais (n) et la probabilité de succès dans chaque essai (p).

 La distribution binomiale peut prendre différentes formes en fonction des valeurs de n et p. Pour de grandes valeurs de n et des valeurs modérées de p, la distribution binomiale est approximativement normale, ce qui signifie qu'elle a une forme de cloche symétrique.

```python
from scipy.stats import binom

x = np.arange(0, 10)
# On tire 10 fois une pièce de monnaie.
# La probabilité d'obtenir face (succès) est de 50%
n, p = 10, 0.5

# Fonction de masse de probabilité (PMF) : 
# Probabilité d'obtenir x succès
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, color = 'blue')
plt.xlabel("Nb de fois qu'on tombe sur face")
plt.ylabel('Probabilité')
plt.title(f'Fonction de masse de probabilité (n = {n}, p = {p})')
plt.show()
```

#### Distribution de Poisson
La distribution de Poisson est une distribution de probabilité discrète qui décrit le nombre d'événements se produisant dans un intervalle de temps ou d'espace fixe, lorsque ces événements se produisent avec une fréquence moyenne constante et indépendamment du temps écoulé depuis le dernier événement. La distribution de Poisson est définie par un seul paramètre, souvent noté λ (lambda), qui représente le nombre moyen d'événements dans l'intervalle donné. Par exemple, si vous observez en moyenne 3 accidents de voiture par jour dans une certaine intersection, λ serait égal à 3. La distribution de Poisson est souvent utilisée pour modéliser des phénomènes tels que le nombre de clients arrivant à un guichet, le nombre de pannes dans un système, le nombre d'appels à un centre d'assistance dans une heure, etc.

```python
from scipy.stats import poisson

# Paramètre lambda pour la distribution de Poisson
lambda_param = 3

# Générer des valeurs de la distribution de Poisson
x = np.arange(0, 11) 
poisson_values = poisson.pmf(x, lambda_param)

# Tracer la distribution de Poisson
plt.figure(figsize=(8, 6))
plt.plot(x, poisson_values, alpha=0.7, color='b')
plt.xlabel('Nombre d\'événements')
plt.ylabel('Probabilité')
plt.title('Distribution de Poisson')
plt.show()
```

### Z-score

Le score Z représente combien d'écart-type sépare une observation d'un jeu de données à la moyenne de celle-ci. L'utilisation du z-score est courante dans les analyses statistiques, en particulier lors de la détection des valeurs extrêmes  dans un ensemble de données. Les points de données ayant un z-score élevé (en valeur absolue) sont généralement considérés comme des valeurs extrêmes et peuvent nécessiter un examen plus approfondi.

Il peut être calculé à l'aide de la formule suivante :

$$ Z = \frac{{X - \mu}}{{\sigma}} $$

Où :
* $ X $ est l'observation.
* $ \mu $ est la moyenne de l'échantillon.
* $ \sigma $ est l'écart type de l'échantillon.




```python
from scipy import stats
data = np.array([6, 7, 7, 12, 13, 13, 15, 16, 19, 22])
stats.zscore(data)
```

```python
# Exemple de calcul
obs=data[0]
mu=data.mean()
sigma=data.std()
z=(obs-mu)/sigma
z
```

### Intervalle de confiance

Un intervalle de confiance est une plage de valeurs dans laquelle on estime qu'un paramètre statistique, comme la moyenne, est susceptible de se trouver, avec un certain niveau de confiance. En d'autres termes, c'est une méthode statistique utilisée pour estimer l'incertitude autour d'une mesure statistique d'une population basée sur un échantillon de celle-ci.

L'intervalle de confiance (IC) à 95 % est un intervalle calculé à partir d'un échantillon de données provenant d'une séquence infinie, dont 95 % comprennent le paramètre de population. À long terme, 95 % de ces intervalles incluent la véritable moyenne.

Plus le niveau de confiance est élevé (par exemple, 95% au lieu de 90%), plus l'intervalle de confiance sera large pour accueillir une plus grande incertitude.

La fonction `t.interval` permet de calculer l'intervalle de confiance du paramètre de la population que l'on cherche à estimer.


```python
# Importer les bibliothèques nécessaires
import numpy as np
from scipy import stats

# Générer un échantillon de données (pour l'exemple, nous utilisons des données aléatoires)
np.random.seed(0)  # Pour rendre l'exemple reproductible
echantillon = np.random.normal(loc=100, scale=20, size=50)  # Moyenne = 100, Écart-type = 20, Taille de l'échantillon = 50

# Calculer la moyenne et l'écart-type de l'échantillon
moyenne_echantillon = np.mean(echantillon)
#ecart_type_echantillon = np.std(echantillon)#, ddof=1)  # ddof=1 pour indiquer l'échantillon (N-1) au lieu de la population N
print(moyenne_echantillon)
# Niveau de confiance (95% dans cet exemple)
niveau_confiance = 0.95

# Calcul de l'intervalle de confiance
intervalle_confiance = stats.t.interval(niveau_confiance, len(echantillon) - 1, loc=moyenne_echantillon, scale=stats.sem(echantillon))

# Afficher l'intervalle de confiance
print(f"Intervalle de confiance à {niveau_confiance * 100}% : {intervalle_confiance}")
```

### Les tests statistiques
Les tests statistiques permettent de valider une hypothèse par rapport à des données. Voici les étapes afin de réaliser un test statistique:

* Construire les hypothèses H0 et H1
* Déterminer le risque d'erreur acceptable 
* Choisir le test adapté
* Calculer la *p-value* et l'interpréter


#### L'hypothèse nulle et l'hypothèse alternative

L'hypothèse nulle (H0) représente le statu quo tandis que l'hypothèse alternative (H1) est l'hypothèse que l'on souhaite démontrer.

Supposons que nous ayons deux médicaments (A et B), et nous voulons savoir si le médicament A est plus efficace que le médicament B en termes de réduction de la pression artérielle. Nous prélevons des échantillons de patients sous chaque traitement et mesurons leur pression artérielle

Voici ce que pourrait être les deux hypothèses:
* Hypothèse Nulle (H0) : Les moyennes de la pression artérielle des patients sous les deux médicaments sont égales.
* Hypothèse Alternative (H1): La moyenne de la pression artérielle des patients sous le médicament A est inférieure à la moyenne de la pression artérielle des patients sous le médicament B.

#### Risque d'erreur

Il existe deux types d'erreur:
* Type 1: On suppose qu’H0 est fausse alors qu’en réalité H0 est vraie 
* Type 2: On suppose qu’H0 est vraie alors qu’en réalité H0 est fausse 

Par défaut l'erreur de type 1 (aussi appelé alpha) est fixée à 5%. Ainsi, on accepte qu'on a 5% de chance de rejeter H0 si elle est vraie.

Dans le cas de notre exemple, cela reviendrait à dire que le traitement A est efficace à faire diminuer la pression artérielle.

L'erreur de type 2 dans notre exemple reviendrait à dire qu'il n'y a pas de différence au niveau de la pression artérielle en fonction du traitement même si le traitement A est en fait plus efficace pour faire diminuer la pression artérielle que le traitement B.



#### Les tests unilatéraux et bilatéraux 
Il existe deux types de tests:
* unilatéraux
* bilatéraux
La différence réside au niveau de l'hypothèse alternative (H1). Dans un test unilatéral, il existe une différence entre les deux groupes dont le sens est connu (comme dans notre exemple). Dans un test bilatéral, il existe une différence entre les deux groupes mais on ne mentionne pas le sens.

#### Choix du test
Il existe beaucoup de tests statistiques différents. Chaque test a ses propres conditions d’application.

Afin de déterminer le test adapté, on a besoin d’identifier certains critères :

* Nature des variables (catégorielle, continues etc.)
* Nombre de groupes
* Normalité de la distribution des variables

Dans les prochaines sections, nous allons couvrir les tests suivants:
* Z-test
* T-test
* Test du Khi-Deux
* ANOVA

#### p-value
Afin de déterminer si on doit rejeter ou non l'hypothèse nulle, nous allons calculer la *p-value* avec le test d'hypothèse choisi. Le calcul est effectué en fonction de la probabilité de distribution c'est-à-dire une fonction qui décrit toute les valeurs possibles et la probabilité de celles-ci dans un certain intervalle de valeurs.  La *p-value* quantifie la probabilité d'observer des données similaires à celles observées, supposant que l'hypothèse nulle est vraie. En d'autres termes, la valeur p indique à quel point les données sont compatibles avec l'hypothèse nulle.

Si la *p-value* est inférieure à alpha (typiquemment 0,05), cela suggère que les données qu'on a observées sont peu probables si l'hypothèse nulle est vraie. Cela signifie qu'il y a suffisamment de preuves pour rejeter l'hypothèse nulle au profit de l'hypothèse alternative. 

Si la *p-value* est supérieure à alpha (typiquemment 0,05),  cela indique que les données qu'on a observées sont assez probables même si l'hypothèse nulle est vraie. Cela signifie qu'il n'y a pas suffisamment de preuves pour rejeter l'hypothèse nulle.

Il est important de noter que la p-value n'indique pas la probabilité que l'hypothèse nulle soit vraie ou fausse. Elle mesure la probabilité d'obtenir les données observées, compte tenu de l'hypothèse nulle.

### Z-Test 
Le Z-test est utilisé pour évaluer les différences entre les moyennes de deux groupes lorsqu'on connaît l'écart-type de la population de notre échantillon de données.

Il existe trois principaux types de T-test:

* Z-test pour échantillon unique (comparaison d'une moyenne avec une moyenne connue)
* Z-test indépendant (2 groupes indépendants)
* Z-test apparié (2 groupes dépendants)

Le test s'effectue en calculant le score Z (légèrement modifié) et en calculant la p-value associée à celle-ci grâce à un tableau de distribution normale.

Le calcul du score Z pour échantillon unique:


$$ Z = \frac{{\bar{X} - \mu_0}}{{\frac{{\sigma}}{{\sqrt{n}}}}} $$

Où :
- $ \bar{X}$ est la moyenne.
- $\mu_0$ est la moyenne de la population
- $ \sigma$ est l'écart-type de la population.
- $n$ est la taille de l'échantillon.


Le calcul du score Z lorsqu'il y a 2 groupes indépendants:

$$ Z = \frac{{\bar{X}_1 - \bar{X}_2}}{{\sqrt{\frac{{\sigma_1^2}}{{n_1}} + \frac{{\sigma_2^2}}{{n_2}}}}} $$

Où :
- $ \bar{X}_1$ et $\bar{X}_2$ sont les moyennes des deux groupes.
- $ \sigma_1^2$ et $\sigma_1^2$ sont les écart-types des populations des 2 groupes.
- $n_1$ et $n_2$ sont les tailles des échantillons des deux groupes.

Le calcul du score Z lorsqu'il y a 2 groupes appariée:


$$Z = \frac{\bar{X}_d}{\frac{\sigma_d}{\sqrt{n}}}$$

Où :
- $ \bar{X}_d$ est la moyenne des différences entre les paires.
- $ \sigma$ est l'écart type des différences entre les paires de la population.
- $n$ est la taille de l'échantillon.



**Exemple** 
Le directeur de l'université XYZ affirme que les étudiants qui étudient dans son établissement sont plus intelligents que la moyenne des écoles. On obtient les scores de QI de 50 étudiants allant à l'université XYZ et la moyenne s'avère être de 110. La moyenne du QI de la population est de 100 et l'écart-type est de 15. Dites si l'affirmation du directeur est correcte ou non avec un risque d'erreur de type 1 de 5 %.

Hypothèses:
* H0: la moyenne de QI des étudiants de XYZ est égale à celle de la population
* H1: la moyenne de QI des étudiantes de XYZ est supérieur à celle de la population

Ici, H1 est unilatéral.

```python
# Information
sample_mean = 110
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05

# Calcul du z-score
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(50))
print('Z-Score :',z_score)

# P-Value 
# H1 est unilatéral et on cherche la probabilité d'être supérieur au z_score
# Si H1 était unilatéral et qu'on cherchait la probabilité d'être inférieur au z_score: stats.norm.cdf(z_score)
# Si H1 était bilatéral: 2 * (1-stats.norm.cdf(np.abs(z_score)))
#CDF:  La fonction de distribution cumulative (cdf) 
# donne la probabilité d'obtenir un z-score inférieur à la valeur passée en argument

p_value = 1-stats.norm.cdf(z_score)

print('p-value :',p_value)

# Décision sur l'hypothèse
if p_value < alpha:
    print("On rejette H0")
else:
    print("On ne rejette pas H0")
```

### T-Test 
Le T-test est également une statistique inférentielle utilisée pour évaluer les différences entre les moyennes de deux groupes.
La différence avec le Z-test est qu'il peut être utilisé lorsqu'on ne connait pas l'écart-type de la population et pour des échantillons qui peuvent être de plus petite taille que le Z-test.

Il existe trois principaux types de T-test:

* T-test pour échantillon unique
* T-test indépendant
* T-test apparié

Le test s'effectue en calculant la valeur *t* puis en comparant celle-ci avec un tableau de distribution t. Nous pouvons ainsi trouver la *p-value* associée à cette t-value.  En fonction du type de test, la formule pour calculer la valeur *t* varie


#### T-test pour échantillon unique

Le T-test pour échantillon unique est utilisé pour comparer la moyenne d’un échantillon à une moyenne standard connue.

Hypothèses:
* H0: La moyenne de l'échantillon est égale à la moyenne standard connue
* H1: La moyenne n'est pas égale à la moyenne standard connue (test bilatéral)

Voici les conditions à remplir pour pouvoir l'utiliser:
* le jeu de données provient d'un échantillon aléatoire
* Les données sont numériques et continues
* Le jeu de données doit avoir une distribution relativement normale ou l'échantillon est suffisamment grand 
* Pas de valeurs extrêmes

La fonction `ttest_1samp()` de la librairie *scipy* permet d'effectuer ce test. L'argument `alternative` nous permet de définir le sens de H1. Par défaut, celui-ci est bilatéral.
Voici le lien de la documentation de la fonction: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_1samp.html

**Exemple**
Supposons que nous menions une étude pour évaluer le temps moyen passé par les étudiants à étudier chaque jour. Nous prélevons un échantillon de 30 étudiants et enregistrons le nombre d'heures qu'ils passent à étudier quotidiennement.

Notre hypothèse nulle (H0) est que le temps moyen passé à étudier est égal à 3 heures par jour, tandis que notre hypothèse alternative (H1) est que le temps moyen passé à étudier n'est pas égal à 3 heures par jour.


```python
data=[3.2,2.8,3.5,2.9,3.1,3.4,3.0,
      3.2,3.3,3.1,2.7,3.6,2.9,3.3,
      3.2,3.4,3.1,3.5,3.2,3.0,3.3,
      3.4,3.1,2.8,3.2,3.0,3.4,3.2,3.1,3.3]
#On valide la distribution de nos données
plt.hist(data, bins=10,
         density=True,#y_value= counts / (sum(counts) * bins_width)
         alpha=0.5,
         color='steelblue')
plt.xlabel('Valeurs')
plt.ylabel('Densité')
plt.title("Distribution du temps moyen d'étude")

t_value,p_value=stats.ttest_1samp(data,3) #où 3 est la moyenne de la population estimé si H0 est vraie
print('p-value pour un test bilatéral %f'%p_value)
# Décision sur l'hypothèse
if p_value < 0.05:
    print("On rejette H0")
else:
    print("On ne rejette pas H0")
```

Si alpha= 5%, on peut rejeter H0. On peut conclure que le temps moyen passé à étudier n'est pas égal à 3 heures par jour.

#### T-test indépendant
Ce test est utilisé pour comparer les moyennes de deux groupes et déterminer s'ils sont statistiquement différents l'un de l'autre.

Hypothèses:
* Hypothèse Nulle (H0) : Les moyennes des deux groupes sont égales
* Hypothèse Alternative (H1): Les moyennes des deux groupes sont différentes.

Voici les conditions à remplir pour pouvoir l'utiliser: 

1. La variable dépendante doit être continue.

2. La variable indépendante doit avoir 2 groupes (catégories).

3. Les observations doivent être indépendantes.

4. Pas de valeurs extrêmes dans la variable dépendante

5. La variable dépendante doit avoir une distribution relativement normale au sein de chacun des groupes de la variable indépendante ou l'échantillon est suffisamment grand.

Il existe 2 versions du T-test indépendant qui peut être utilisé en fonction de la condition suivante:
Si les écarts-types des valeurs de la variable dépendante sont relativement égale pour les deux groupes, on peut utiliser le **Student T-test**. Sinon, on peut utiliser le **Welch T-test**.

La fonction `ttest_ind` de la librairie *scipy* permet d'effectuer ce test. L'argument `alternative` nous permet de définir le sens de H1. Par défaut, celui-ci est bilatéral. L'argument `equal_var`nous permet de spécifier si on veut effectuer le Student T-test (`equal_var=True`) ou le Welch T-test (`equal_var=False`). Par défaut, `equal_var` est égal à `True`.
Voici le lien de la documentation de la fonction: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html

**Exemple**
Nous avons effectué un test sur une rizière, sur une partie on a traité les plants avec du chlorure d'ammonium et l'autre avec de l'urée. Nous souhaitons analyser les rendements en grain afin de déterminer si l'effet des traitements est le même ou non.

* H0: le chlorure d'ammonium et l'urée ont le même impact sur le rendement du grain
* H1: le chlorure d'ammonium et l'urée n'ont pas le même impact sur le rendement du grain

```python
#mean = np.random.uniform(11, 16)  # Generate a random mean between 11 and 16
#std_dev = 1

Ammonium_chloride=[12.5, 15.4, 15.2, 14.2, 11.9, 12.9, 13.9, 16.8, 12.7,  9.5, 13.9,
       15. , 12.2, 13.2, 11.9, 11.7, 11.8, 14. , 14.5, 13.6, 15.3, 11.5,
       13. , 14.6, 16.1, 14.5, 11.8, 15.1, 12.1]

Urea=[12.9, 12.7, 12.9, 15.9, 11.2, 12. , 14. , 13.1, 13.5, 11.9, 13.8,
       13.6, 13.7, 13.8, 15.6, 14.6, 11. , 14.7, 12.3, 14.4, 13.6, 12.8,
       12.2, 12.5, 15.8, 15.5, 13.3, 13.5, 13.3]
from scipy import stats

#On valide la distribution de nos 2 groupes
fig, ax = plt.subplots(1, 2, figsize=(10, 4)) 
ax[0].hist(Ammonium_chloride, bins=5,
         density=True,#y_value= counts / (sum(counts) * bins_width)
         alpha=0.5,
         color='steelblue')
ax[0].set_title("Distribution (chlorure d'ammonium)")
ax[1].hist(Urea, bins=5,
         density=True,#y_value= counts / (sum(counts) * bins_width)
         alpha=0.5,
         color='steelblue')
ax[1].set_title("Distribution (urée)")
# On valide les variances
print("Variance des rendements avec chlorure d'ammonium",np.std(Ammonium_chloride))
print("Variance des rendements avec chlorure d'ammonium",np.std(Urea))
# Si ratio des variances (max/min) est inférieure à 2 ou 3, on peut garder le student T-test
# 2 ou 3 est une règle du pouce
print("Ratio des variance",np.std(Urea)/np.std(Ammonium_chloride))

# La t-value mesure l'écart entre les moyennes de deux groupes 
# exprimé en termes d'écart-type de l'échantillon.
t_value,p_value=stats.ttest_ind(Ammonium_chloride,Urea)

print('p-value pour un test bilatéral %f'%p_value)
# Décision sur l'hypothèse
if p_value < 0.05:
    print("On rejette H0")
else:
    print("On ne rejette pas H0")
```

Si alpha= 5%, on ne peut pas rejeter H0.

#### T-test apparié
Le test t apparié est utilisé pour comparer les moyennes de deux groupes appariés ou dépendants, comme des mesures avant et après un traitement sur le même groupe de personnes.

Hypothèse Nulle (H0) : Les moyennes des deux groupes appariés sont égales.

Hypothèse Alternative (H1) : Les moyennes des deux groupes appariés sont différentes.

Voici les conditions à remplir pour pouvoir l'utiliser:

1. La variable dépendante doit être continue.

2. La variable indépendante doit avoir 2 groupes (catégories).

3. Les observations doivent être indépendantes.

4. Pas de valeurs extrêmes dans la variable dépendante

5. La variable dépendante doit avoir une distribution relativement normale au sein de chacun des groupes de la variable indépendante ou l'échantillon est suffisamment grand.

6. Les données doivent être appariées, ce qui signifie qu'il y a une correspondance entre les observations des deux groupes. Par exemple, les mesures prises avant et après un traitement pour chaque individu constituent des données appariées.

7. Variances des valeurs de la variable dépendante sont relativement égale pour les deux groupes

La fonction `ttest_rel` de la librairie *scipy* permet d'effectuer ce test. L'argument `alternative` nous permet de définir le sens de H1. Par défaut, celui-ci est bilatéral.
Voici le lien de la documentation de la fonction: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html

**Exemple**
Supposons que nous avons mesuré le poids de 10 personnes avant et après un programme d'exercices, et nous voulons déterminer si le programme d'exercices a eu un impact significatif sur leur poids. 

Hypothèse:
* H0: Les moyennes des poids avant et après sont égales.
* H1: Les moyennes des poids avant et après ne sont pas égales.

```python
from scipy.stats import ttest_rel

# Données de poids avant le programme (en kg)
poids_avant = [70, 75, 72, 68, 71, 74, 69, 73, 72, 70]

# Données de poids après le programme (en kg)
poids_apres = [68, 72, 70, 65, 69, 71, 66, 70, 68, 67]

#On valide la distribution de nos 2 groupes
fig, ax = plt.subplots(1, 2, figsize=(10, 4)) 
ax[0].hist(poids_avant, bins=5,
         density=True,#y_value= counts / (sum(counts) * bins_width)
         alpha=0.5,
         color='steelblue')
ax[0].set_title("Distribution (poids_avant)")
ax[1].hist(poids_apres, bins=5,
         density=True,#y_value= counts / (sum(counts) * bins_width)
         alpha=0.5,
         color='steelblue')
ax[1].set_title("Distribution (poids_apres)")
# On valide les variances
print("Variance des poids_avant",np.std(poids_avant))
print("Variance des poids_apres",np.std(poids_apres))
# Si ratio des variances est inférieure à 2, on peut garder faire le t-test
# 2 ou 3 est une règle du pouce
print("Ratio des variance",np.std(poids_avant)/np.std(poids_apres))

# Effectuer le t-test apparié
t_statistic, p_value = ttest_rel(poids_avant, poids_apres)

# Afficher les résultats
print("Valeur de p :", p_value)

# Interprétation des résultats
alpha = 0.05 
if p_value < alpha:
    print("Les poids avant et après le programme sont significativement différents.")
else:
    print("Il n'y a pas de différence significative entre les poids avant et après le programme.")
```

### Z-test ou T-test
Voici des règles du pouce afin de déterminer quel test prendre:

1. Si vous disposez d'un grand échantillon avec des données normalement distribuées, utilisez le test Z.
2. Si vous avez un petit échantillon et que la population est normale :
    * Utilisez le test Z si vous connaissez l'écart-type de la population.
    * Utilisez le test t si vous ne connaissez pas l'écart-type de la population.

En réalité, il est souvent rare que nous connaissons l'écart-type de la population. Pour cette raison, le T-test est souvent préféré au Z-test.

### Tests du Khi-Deux (chi-square) et interprétations

#### Test d'indépendance du Khi-deux
Le test d'indépendance du Khi-deux (χ²) est utilisé pour analyser s'il existe une association significative entre deux variables catégorielles.

Hypothèses :

* Hypothèse nulle (H0): les variables sont indépendantes, il n'y a pas de relation entre les deux variables catégorielles. Connaître la valeur d'une variable n'aide pas à prédire la valeur de l'autre variable
* Hypothèse alternative (H1): les variables sont dépendantes, il existe une relation entre les deux variables catégorielles. Connaître la valeur d'une variable permet de prédire la valeur de l'autre variable

Le test d'indépendance du Khi-deux fonctionne en comparant les fréquences observées dans l'échantillon (table de contingence) aux fréquences attendues s'il n'y avait pas de relation entre les deux variables catégorielles (donc les fréquences attendues si l'hypothèse nulle était vraie).

La formule pour calculer la valeur de khi-deux est:
$$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$
Où:
* $O_i$ est la fréquence observée pour chaque cellule de la table de contingence
* $E_i$ est la fréquence attendue pour chaque cellule de la table de contingence


La fréquence attendue se calcule ainsi:
$$ E_i = \frac{( \text{Total de la ligne} \times \text{Total de la colonne} )}{\text{Total de l'échantillon}}$$


Le nombre de données doit être suffisamment grand pour que les valeurs des fréquence attendues dans la table de contingence soient supérieures à un certain seuil (souvent 5)

On peut effectuer ce test avec la fonction `chi2_contingency`. Voici le lien vers sa documentation:https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html

Ci-dessous se trouve un exemple de test du Khi-deux entre les variables *cylinders* et *origin*.

Dans cet exemple, nous allons utiliser le jeu de données **auto_mpg**. Les données regroupent différentes informations concernant des voitures:

* **mpg**: distance en miles par gallon (variable continue)
* **cylinders**: nombre de cylindres (discrète)
* **displacement**: cylindrée du moteur (continue)
* **horsepower**: puissance (continue)
* **weight**: poids de l’automobile (continue)
* **acceleration**: Temps en secondes pour aller de 0 à 100km/h (continue)
* **model year**: année du modèle (continue)
* **origin**: continent d’origine de l’automobile (discrète)
* **car name**: nom de la voiture (unique pour chaque observation / discrète)

```python
# Importation du jeu de données : on ajoute le paramètre na_values pour indiquer que les "?"
# dans la variable horsepower sont des valeurs manquantes

auto_mpg = pd.read_csv('auto-mpg.csv',sep = ",",na_values = ["?"])

# On retire les valeurs manquantes
auto_mpg_no_na = auto_mpg.copy()
auto_mpg_no_na = auto_mpg_no_na.dropna()

# On convertit la variable origin en variable catégorielle
auto_mpg_no_na['origin'] = auto_mpg_no_na['origin'].astype('object')
```

```python
# Création de la table de contingence entre les variables cylinders et origin

tab_contingence = pd.crosstab(auto_mpg_no_na.cylinders, auto_mpg_no_na.origin)

tab_contingence
```

```python
# Exemple de calcul de fréquence attendue pour la cellule qui correspond à origin=1 et cylinders=3

ligne_tot=tab_contingence.loc[3,:].sum()
col_tot=tab_contingence.loc[:,1].sum()
tot_ech=tab_contingence.values.sum()

(ligne_tot*col_tot)/tot_ech
```

Dans notre contexte, rejeter l’hypothèse nulle pour le test d’indépendance du Khi-deux signiﬁe qu’il existe une relation signiﬁcative entre le nombre de cylindre et l’origine de l'automobile. 

#### Test d'adéquation du Khi-deux
Le test d'adéquation du khi-deux (χ²) est utilisé pour déterminer si un échantillon de données suit une distribution théorique spécifique. Il compare les fréquences observées dans l'échantillon aux fréquences attendues sous l'hypothèse que les données suivent la distribution. Si la variable est continue on peut la séparée en intervalle (binning) afin d'utiliser ce test.

Hypothèses:
* Hypothèse Nulle (H0) : Les données suivent la distribution théorique spécifiée.
* Hypothèse Alternative (H1) : Les données ne suivent pas la distribution théorique spécifiée.

On peut utiliser la fonction `chisquare` qui nous permet de spécifier la fréquence observée et attendue. Voici le lien vers la documentation:https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chisquare.html

**Exemple**
Un propriétaire de magasin affirme qu'il y un nombre égal de client qui fréquente son magasin à tous les jours de la semaine. Afin de valider cette hypothèse, vous avez mesurer le nombre de client fréquentant sont magasin durant une semaine. 

Vous avez obtenu les résultats suivants:

* Lundi: 50 clients
* Mardi: 60 clients
* Mercredi: 40 clients
* Jeudi: 47 clients
* Vendredi: 53 clients

Les hypothèses sont les suivantes:
* H0: Il y a un nombre égal de client à tous les jours de la semaine
* H1: Il n'y a pas un nombre égal de client à tous les jours de la semaine

```python
# importing packages 
import scipy.stats as stats 
import numpy as np 

expected = [50, 50, 50, 50, 50]
observed = [50, 60, 40, 47, 53]

chi_square_test_statistic, p_value = stats.chisquare( 
observed, expected) 

print('p_value : ' + str(p_value)) 
```

Puisque la *p-value* n'est pas inférieure à 0,05, on ne peut pas rejeter l'hypothèse nulle. On ne peut pas affirmer qu'il n'y a pas un nombre égal de client à tous les jours dans le magasin.

### Le test ANOVA et interprétations

L’utilisation du test statistique ANOVA permet d’estimer la force d’association une variable numérique continue et une variable catégorielle. Ce test est utile quand il y a plus de 2 groupes dans la variable catégorielle.

Ce test compare la moyenne observée de la variable continue au sein de chacun des groupes (catégories) de la variable catégorielle.

Voici les deux hypothèses du test:
* H0: Toutes les moyennes des groupes sont égales
* H1: Au moins, une moyenne est différente des autres

Dans l'exemple ci-dessous, nous allons effectuer le test ANOVA pour valider s'il y a une association entre la variable catégorielle **origin** et la variable continue **mpg**. Les points rouge indique la moyenne de mpg pour chaque catégorie. Le test ANOVA va vérifier si la différence de moyenne de mpg est significativement différente entre les catégories de la variable **origin**.

Le boxplot ci-dessous montre visuellement la variable mpg en fonction de l'origine.

```python
# mpg en fonction du poids
sns.boxplot(data = auto_mpg_no_na,
            x = 'origin',
            y = 'mpg',
            showmeans = True,
            meanprops={"markerfacecolor":"red", "markeredgecolor":"red"}).set_title("boxplot de mpg pour différentes catégories d'origine")
plt.show()
```

```python
import scipy.stats as stats

stats.f_oneway(auto_mpg_no_na['mpg'][auto_mpg_no_na['origin'] == 1],
               auto_mpg_no_na['mpg'][auto_mpg_no_na['origin'] == 2],
               auto_mpg_no_na['mpg'][auto_mpg_no_na['origin'] == 3])
```

La valeur pvalue est <0.05, on peut rejeter l'hypothèse H0. Ainsi la moyenne de la variable mpg pour au moins une catégorie de **origin** est significativement différente. Il y a donc une association significative entre la variable **mpg** et les catégories de **origin**.

### Les corrélations (Pearson, Spearman) et interprétations

#### Corrélation de Pearson
La corrélation de Pearson est une relation linéaire entre deux variables continues telle que les variations de leurs valeurs soient toujours de même sens (corrélation positive) ou de sens opposé (corrélation négative).

Le coefficient de corrélation nommé “r” sert à mesurer la force de la relation linéaire qui existe entre deux variable : à quel point les observations sont alignées autour d’une droite.

Une corrélation proche de 0 indique seulement qu’il n’y a pas de relation linéaire, mais il peut y avoir une relation quadratique.

Les conditions pour utiliser ce test:
* Les deux variables que vous comparez doivent suivrent une distribution normale
* Il n'y a pas présence de valeurs extrêmes

Les hypothèses de ce test sont:
* H0: Il n'y a pas de relation linéaire entre les 2 variables
* H1: Il y a une relation linéaire entre les 2 variables

Pour calculer r et la p_value associée, on peut utiliser la fonction `pearsonr()`. Voici le lien vers la documentation associée:https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html


```python
rho, p_value =stats.pearsonr(auto_mpg_no_na['mpg'], auto_mpg_no_na['cylinders'])
print("Pearson correlation:", rho)
print("p-value",p_value)
```

#### Corrélation de Spearman
La corrélation de Spearman est une mesure statistique qui évalue la force et la direction de la relation entre deux variables. Contrairement à la corrélation de Pearson, qui mesure la corrélation linéaire entre deux variables continues, la corrélation de Spearman évalue la corrélation monotone, c'est-à-dire toute relation systématique, croissante ou décroissante, entre les valeurs des deux variables. Cette mesure peut être utile quand il n'existe pas de relation linéaire entre les variables.

Ce test est utile, entre autres, lorsque vous voulez comparer des variables et que l'une d'entre elles est ordinales (il y a un ordre dans les valeurs (ex: 1-faible, 2-moyen, 3-élevé)). Vous pouvez utiliser ce test lorsqu'il y a présence de valeurs extrêmes et que vos variables ne sont pas distribuées normalement.

Les hypothèses de ce test sont:
* H0: Il n'y a pas de relation monotone entre les 2 variables
* H1: Il y a une relation monotone entre les 2 variables

Pour calculer r et la p_value associée, on peut utiliser la fonction `spearmanr()`. Voici le lien vers la documentation: 
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html

```python

rho, p_value = stats.spearmanr(auto_mpg_no_na['mpg'], auto_mpg_no_na['cylinders'])
 
print("Spearman correlation:", rho)
print("p-value",p_value)
```
